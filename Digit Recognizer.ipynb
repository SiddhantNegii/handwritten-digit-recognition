```python
import numpy as np
import pandas as pd
import tensorflow as tf 
from tensorflow import keras 
from keras.models import Sequential
from keras.layers import Dense
from keras.activations import linear, relu, sigmoid
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
ans = "label"
X_train = df[df.columns.drop(ans)]
Y_train = df[[ans]].copy()
print("The X Training set:")
print(X_train)
print("\n")
print("The Y Training set:")
print(Y_train)
```

```python
tf.random.set_seed(1234)
model = Sequential([
    tf.keras.Input(shape=(784,)),
    Dense(units=25, activation='relu', name="layer1"),
    Dense(units=15, activation='relu', name="layer2"),
    Dense(units=10, activation='linear', name="layer3"),
  ], name = "my_model"
)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import numpy as np

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train.reshape(-1, 28, 28, 1).astype("float32") / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype("float32") / 255.0

model = Sequential([
    tf.keras.Input(shape=(28, 28, 1)),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='linear')
])

model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy']
)

early_stopping_callback = EarlyStopping(
    monitor='val_accuracy', 
    patience=5,
    restore_best_weights=True
)
lr_scheduler_callback = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.5,
    patience=3
)
checkpoint_callback = ModelCheckpoint(
    filepath='my_best_model.keras',
    monitor='val_accuracy',
    save_best_only=True
)

print("Starting CNN model training with advanced callbacks...")
history = model.fit(
    x_train, 
    y_train, 
    epochs=100,
    batch_size=128, 
    validation_split=0.1,
    callbacks=[early_stopping_callback, lr_scheduler_callback, checkpoint_callback]
)
best_model = keras.models.load_model("my_best_model.keras")
best_model.evaluate(x_test, y_test)
best_model.save("my_model.keras")
```

```python
import tkinter as tk
from PIL import Image, ImageDraw, ImageOps
import numpy as np
from tensorflow import keras
import tensorflow as tf

class DigitRecognizer:
    def __init__(self, model_path):
        self.model = keras.models.load_model(model_path, compile=False)
        self.window = tk.Tk()
        self.window.title("Draw a Digit")

        self.canvas_size = 280
        self.canvas = tk.Canvas(self.window, width=self.canvas_size, height=self.canvas_size, bg='white')
        self.canvas.pack()

        self.canvas.bind("<B1-Motion>", self.paint)
        self.canvas.bind("<ButtonRelease-1>", self.reset)

        self.label = tk.Label(self.window, text="Draw a digit", font=("Helvetica", 16))
        self.label.pack()

        tk.Button(self.window, text="Predict", command=self.predict).pack()
        tk.Button(self.window, text="Clear", command=self.clear).pack()

        self.image = Image.new("L", (self.canvas_size, self.canvas_size), 255)
        self.draw = ImageDraw.Draw(self.image)

        self.last_x, self.last_y = None, None 

    def paint(self, event):
        x, y = event.x, event.y
        r = 7 

        if self.last_x is not None and self.last_y is not None:
            self.canvas.create_line(self.last_x, self.last_y, x, y,
                                    width=r*2, fill='black',
                                    capstyle=tk.ROUND, smooth=True)
            self.draw.line([self.last_x, self.last_y, x, y], fill=0, width=r*2)

        self.last_x, self.last_y = x, y

    def reset(self, event):
        self.last_x, self.last_y = None, None


    def clear(self):
        self.canvas.delete("all")
        self.draw.rectangle([0, 0, self.canvas_size, self.canvas_size], fill=255)
        self.label.config(text="Draw a digit")

    def preprocess(self):
        inverted_image = ImageOps.invert(self.image)
        
        bbox = inverted_image.getbbox()
        
        if bbox is None:
            return np.zeros((1, 28, 28, 1), dtype=np.float32)

        cropped_image = inverted_image.crop(bbox)

        width, height = cropped_image.size
        new_size = max(width, height)

        padded_image = Image.new("L", (new_size, new_size), 0)
        padded_image.paste(cropped_image, ((new_size - width) // 2, (new_size - height) // 2))

        resized_image = padded_image.resize((20, 20), Image.Resampling.LANCZOS)

        final_image = Image.new("L", (28, 28), 0)
        final_image.paste(resized_image, (4, 4)) 

        img_array = np.array(final_image).astype("float32") / 255.0

        return img_array.reshape(1, 28, 28, 1)

    def predict(self):
        img = self.preprocess()
        preprocessed_img = Image.fromarray((img[0] * 255).astype(np.uint8).reshape(28, 28))
        preprocessed_img.save("preprocessed_digit.png")
        
        logits = self.model.predict(img)
        print("Raw logits (1x10 array):", logits)
        print("Probabilities (softmax output):", tf.nn.softmax(logits).numpy())
        
        probs = tf.nn.softmax(logits).numpy()
        digit = np.argmax(probs)
        confidence = probs[0][digit] * 100
        self.label.config(text=f"Prediction: {digit} ({confidence:.2f}%)")

    def run(self):
        self.window.mainloop()
```

```python
app = DigitRecognizer("my_model.keras")
app.run()
```
